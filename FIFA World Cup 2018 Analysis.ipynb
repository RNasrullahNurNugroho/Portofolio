###### 17.	Membuat minimal 3 Visualisasi data yang dikehendaki
    
    a. Visualisasi Negara Berdasarkan Total Goal.
    b. Visualisasi Negara Berdasarkan Total Passing.
    c. Visualisasi Negara Berdasarkan Penguasaan Bola.

```python
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
DF = pd.DataFrame(Data_Match[['Team', 'Continent','Ball Possession %', 'Goal Scored', 'Passes', 'Passes Completed']])
DF.head()
```

```python
df_1 = pd.DataFrame(DF.groupby(['Team', 'Continent'])['Ball Possession %'].mean()).reset_index()
df_1.head()
```

```python
df = pd.DataFrame(DF.groupby(['Team', 'Continent'])['Goal Scored', 'Passes', 'Passes Completed'].sum()).reset_index()
df.head()
```

```python
df = pd.DataFrame(df.merge(df_1, on= ['Team','Continent'], how='left')).reset_index(drop=True)
df.head()
```

```python
df = df.sort_values('Goal Scored', ascending=False).reset_index(drop=True)
df.head()
```

a. Visualisasi Negara Berdasarkan Goal Scored

```python
df = df.sort_values('Goal Scored', ascending=False).reset_index()
df = df.set_index('Team')
plt.figure(figsize=(12,12))
df['Goal Scored'].plot(kind = 'barh')
plt.title('Distribusi Total Goal Scored menurut Negara')
plt.xlabel('Total Goal')
plt.ylabel ('Nation')
plt.show()
```

b. Visualisasi Negara Berdasarkan Passing

```python
df = df.sort_values('Passes', ascending=False).reset_index()
df = df.set_index('Team')
plt.figure(figsize=(12,12))
df['Passes'].plot(kind = 'barh')
plt.title('Distribusi Passing menurut Negara')
plt.xlabel('Total Passing')
plt.ylabel ('Nation')
plt.show()
```

c. Visualisasi Negara Berdasarkan Penguasaan Bola

```python
df = df.sort_values('Ball Possession %', ascending=False).reset_index()
df = df.set_index('Team')
plt.figure(figsize=(12,12))
df['Ball Possession %'].plot(kind = 'barh')
plt.title('Distribusi Ball Possession menurut Negara')
plt.xlabel('Rata-Rata Penguasaan Bola')
plt.ylabel ('Nation')
plt.show()
```
